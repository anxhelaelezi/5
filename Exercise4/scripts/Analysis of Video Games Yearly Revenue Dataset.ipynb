## Analysis of Video Games Yearly Revenue Dataset

This Jupyter Notebook analyzes trends in video game revenue, ownership, and release patterns over time. This dataset includes yearly information on total estimated revenue, estimated game owners, and the number of games released, and is used to uncover key insights into the video game industry's growth and dynamics.


```python
# Imports and Preprocessing
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os

# Define the relative path to the data folder
file_path = os.path.join(os.getcwd(), "../data", "games_yearly_revenue.csv")

# Read the CSV file
games_yearly_revenue = pd.read_csv(file_path)
print(games_yearly_revenue.head())
```

### Yearly Trend of Total Estimated Revenue

```python
# Line Plot 1: Yearly trend of total estimated revenue over time
plt.figure(figsize=(14, 5))
sns.lineplot(x='Release_Year', y='Total_Estimated_Revenue', data=games_yearly_revenue, marker='o')
plt.title('Yearly Trend of Total Estimated Revenue')
plt.xlabel('Release Year')
plt.ylabel('Total Estimated Revenue')
plt.show()
```

This plot shows a significant fluctuation in revenue over the years, with certain periods showing sharp increases, indicating successful game releases or industry growth during those times.

### Yearly Trend of Total Estimated Owners

```python
# Line Plot 2: Yearly trend of total estimated owners over time
plt.figure(figsize=(14, 5))
sns.lineplot(x='Release_Year', y='Total_Estimated_Owners', data=games_yearly_revenue, marker='o')
plt.title('Yearly Trend of Total Estimated Owners')
plt.xlabel('Release Year')
plt.ylabel('Total Estimated Owners')
plt.show()
```

The number of game owners also fluctuates, mirroring the revenue trends, which suggests a strong correlation between user acquisition and revenue growth.

### Yearly Trend of Number of Games Released

```python
# Line Plot 3: Yearly trend of the number of games released
plt.figure(figsize=(14, 5))
sns.lineplot(x='Release_Year', y='Number_of_Games', data=games_yearly_revenue, marker='o')
plt.title('Yearly Trend of Number of Games Released')
plt.xlabel('Release Year')
plt.ylabel('Number of Games')
plt.show()
```

The number of games released annually shows a positive upward trend over time, suggesting a growing gaming industry.

### Revenue Contribution by Year

```python
# Calculate the total revenue
total_revenue = games_yearly_revenue['Total_Estimated_Revenue'].sum()

# Define a threshold percentage (e.g., 2%)
threshold = 2
threshold_value = (threshold / 100) * total_revenue

# Group values below the threshold into "Others"
# Filter values above the threshold
above_threshold = games_yearly_revenue[games_yearly_revenue['Total_Estimated_Revenue'] >= threshold_value]

# Sum the values below the threshold to create the "Others" category
below_threshold_sum = games_yearly_revenue[games_yearly_revenue['Total_Estimated_Revenue'] < threshold_value]['Total_Estimated_Revenue'].sum()

# Append "Others" if there are values below the threshold
if below_threshold_sum > 0:
    above_threshold = pd.concat([above_threshold, pd.DataFrame({'Release_Year': ['Others'], 'Total_Estimated_Revenue': [below_threshold_sum]})])

# Plot the pie chart: Revenue Contribution by Year
plt.figure(figsize=(6, 6))
plt.pie(above_threshold['Total_Estimated_Revenue'], labels=above_threshold['Release_Year'], autopct='%1.1f%%', startangle=140)
plt.title('Revenue Contribution by Year (Grouped)')
plt.tight_layout()
plt.show()
```

This chart displays the major contributors to total revenue, with smaller contributions grouped under "Others".

Revenue is highly concentrated in a few key years. This concentration suggests potential peaks in market activity, possibly influenced by major game launches or economic factors.

### Owners Distribution by Year

```python
# Pie Chart 2: Owners Distribution by Year (Grouped)
# Calculate the total owners
total_owners = games_yearly_revenue['Total_Estimated_Owners'].sum()
threshold_value_owners = (threshold / 100) * total_owners

# Group values below the threshold into "Others"
above_threshold_owners = games_yearly_revenue[games_yearly_revenue['Total_Estimated_Owners'] >= threshold_value_owners]
below_threshold_sum_owners = games_yearly_revenue[games_yearly_revenue['Total_Estimated_Owners'] < threshold_value_owners]['Total_Estimated_Owners'].sum()

if below_threshold_sum_owners > 0:
    above_threshold_owners = pd.concat([above_threshold_owners, pd.DataFrame({'Release_Year': ['Others'], 'Total_Estimated_Owners': [below_threshold_sum_owners]})])

# Plot the pie chart
plt.figure(figsize=(6, 6))
plt.pie(above_threshold_owners['Total_Estimated_Owners'], labels=above_threshold_owners['Release_Year'], autopct='%1.1f%%', startangle=140)
plt.title('Owners Distribution by Year (Grouped)')
plt.tight_layout()
plt.show()
```

This chart shows the distribution of game owners across years, with minor contributions also grouped as "Others".

### Games Released Distribution by Year

```python
# Pie Chart 3: Games Released Distribution by Year (Grouped)
# Calculate the total number of games
total_games = games_yearly_revenue['Number_of_Games'].sum()
threshold_value_games = (threshold / 100) * total_games

# Group values below the threshold into "Others"
above_threshold_games = games_yearly_revenue[games_yearly_revenue['Number_of_Games'] >= threshold_value_games]
below_threshold_sum_games = games_yearly_revenue[games_yearly_revenue['Number_of_Games'] < threshold_value_games]['Number_of_Games'].sum()

if below_threshold_sum_games > 0:
    above_threshold_games = pd.concat([above_threshold_games, pd.DataFrame({'Release_Year': ['Others'], 'Number_of_Games': [below_threshold_sum_games]})])

# Plot the pie chart
plt.figure(figsize=(6, 6))
plt.pie(above_threshold_games['Number_of_Games'], labels=above_threshold_games['Release_Year'], autopct='%1.1f%%', startangle=140)
plt.title('Games Released Distribution by Year (Grouped)')
plt.tight_layout()
plt.show()
```

This chart highlights the share of game releases per year, grouping years with smaller numbers under "Others".

The release distribution indicates that a few years contributed a larger volume of game releases, with the “Others” category containing most of the lesser-release years. This pattern implies specific periods with heightened development activity, potentially due to industry trends or technological advancements.
